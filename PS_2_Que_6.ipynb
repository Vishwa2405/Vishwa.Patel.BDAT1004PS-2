Implement a class named HeadingParser that can be used to parse an HTML
document, and retrieve and print all the headings in the document. <br>You should implement your class as a subclass of HTMLParser, defined in Standard Library module html.parser. When fed a string containing HTML code, your class should print the headings, one per line and in the order in which they appear in the document. Each heading should be indented as follows: an h1 heading should have indentation 0, and h2 heading should have indentation 1, etc. Test your implementation using w3c.html

```python
from html.parser import HTMLParser

# subclass of HTMLParser
class HeadingParser(HTMLParser):
    def __init__(self):
        super().__init__()
        self.capture = False
        self.indent = 0

    def handleStarttag(self, tag, attrs):
        if tag in ["h1", "h2", "h3", "h4", "h5", "h6"]:
            self.capture = True
            self.indent = int(tag[1]) - 1  # Setting the indentation level

    def handleEndtag(self, tag):
        if tag in ["h1", "h2", "h3", "h4", "h5", "h6"]:
            self.capture = False

    def handle_data(self, data):
        if self.capture:
            indentation = ' ' * self.indent  # Multiplying indentation level
            print(f"{' ' * self.indent}{data}")

with open('/content/w3c.txt', 'r') as infile:
    content = infile.read()

hp = HeadingParser()
hp.feed(content)
```
