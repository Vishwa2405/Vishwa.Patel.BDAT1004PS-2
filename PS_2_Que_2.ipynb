```python
# Function file_length
def fileLength(name):
    try:
        with open(name, 'r') as file:
            txt = file.read()

        # length of file content
        print(len(txt))

    except FileNotFoundError:
        # friendly message
        print(f"File {name} not found.")

if __name__ == "__main__":
    fileLength('/content/midterm.py')

    fileLength('idterm.py')
```
