```python
words = ['The', 'quick', 'brown', 'fox', 'jumps', 'over', 'the', 'lazy', 'dog']
```

```python
uppercaseWords = [word.upper() for word in words]
uppercaseWords
```

```python
lowercaseWords = [word.lower() for word in words]
lowercaseWords
```

```python
lengthWords = [len(word) for word in words]
lengthWords
```

```python
detailedWords = [[word.upper(), word.lower(), len(word)] for word in words]
detailedWords
```

```python
words4OrMoreCharacters = [word for word in words if len(word) >= 4]
words4OrMoreCharacters
```
